### Scaling Features `X_train`

```python
X_train.head(2)
```

```python
scaled_features = ['bathrooms',
                   'bedrooms',
                   'price',
                   'num_photos',
                   'num_features',
                   'description_length']
```

```python
for feature in scaled_features:
    mean = X_train[feature].mean()
    std_dev = X_train[feature].std()
    
    X_train[feature] = X_train[feature].apply(lambda x: (x - mean)/std_dev)
```
